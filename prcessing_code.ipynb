```python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt

lenna1=img.imread('/home/kim/PycharmProjects/tensorflow2.0-gpu/donghwikim/imageclass/lenna.JPG')
lenna=lenna1[:,:,0]
print(lenna.shape)
mask_size=5
blu = np.ones((mask_size,mask_size)) / mask_size**2

out=np.copy(lenna)
for i in range(lenna.shape[0]):
    for j in range(lenna.shape[1]):
        try:
            out[i,j]=np.sum(blu*lenna[i:i+mask_size,j:j+mask_size])
        except ValueError:
            continue

plt.imshow(lenna1,cmap='gray')
plt.show()
plt.imshow(out,cmap='gray')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img

lenna1=img.imread('/home/kim/PycharmProjects/tensorflow2.0-gpu/donghwikim/imageclass/lenna.JPG')
lenna=lenna1[:,:,0]

mask_size=5
blu = np.ones((5,5)) / 5**2
size=int((mask_size-1) / 2)

out=np.copy(lenna)
print(lenna.shape)
for i in range(lenna.shape[0]):
    for j in range(lenna.shape[1]):
        try:
            out[i,j]=np.sum(blu*lenna[i-size,j:j])
        except ValueError:
            continue
plt.imshow(lenna1,cmap='gray')
plt.show()
plt.imshow(out,cmap='gray')
plt.show()
```
